```python
"""Seaborn allows you to plot multiple plots on the same page across multiple input variables.
By default, it chooses the colors and styles for you but allows for fine-tuning to get things the way you like.
Many of the plots perform statistical estimation and plot graphically to help understand confidence intervals.
With Seaborn, its very easy to switch between plot types and styles with the kind argument.
It does not support 3D graphics and does not do as well with large datasets.  Its also not as good for advanced graphs
"""
```

```python
#importing the libraries I may be using
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('diabetic_data.csv')
# Only keep the main genders and races
slimdf = df[df['gender'].isin(['Male', 'Female']) & df['race'].isin(['Caucasian', 'AfricanAmerican', 'Hispanic'])]

#want to see all the columns and first 5 obs
pd.set_option('display.max_columns', None)
print(slimdf.head(5))

# Create a scatter plot
#sns.scatterplot(data=slimdf, x='race', y='time_in_hospital')
#plt.show()

# See a line plot with gender, race and some numeric vars of interest
sns.relplot(
    data=slimdf, kind="line",
    x="time_in_hospital", y="num_lab_procedures", col="gender",
    hue="race", style="race",
)
```

```python
#sampling down the dataframe to reduce the processing and crowdedness on the plots
sampled_df = slimdf.sample(frac=0.05, random_state=222)
#do a pair plot on the numeric variables and on the readmitted variable
sns.pairplot(sampled_df, vars=['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'number_diagnoses'], hue="readmitted")
```
